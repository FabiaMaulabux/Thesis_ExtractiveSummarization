```
def log_as_csv(logname):
  with open(logname, newline='') as csvfile:
    metrics = list(csv.reader(csvfile))
  return metrics
```

```
def rouge_csv(rouge_array):
  with open('ROUGE.csv', 'w') as fw:
      for i in range(len(rouge_array)):
        fw.write(rouge_array[i])
        fw.write('\n')
  fw.close()
  return "ROUGE.csv"
```

```
def save_model_rf(runname, modelname, model, metrics, log, rouge_array):
  csv = rouge_csv(rouge_array)
  with mlflow.start_run(experiment_id=2860077901755569, run_name=runname) as run:
    mlflow.sklearn.log_model(model, modelname)
    mlflow.log_artifact(log, artifact_path =None)
    mlflow.log_artifact(csv, artifact_path =None)

    values = np.array(metrics[1]).astype(np.float)
    mlflow.log_param('Epochs', str(int(metrics[len(metrics)-1][0]) + 1))
    for i in range(len(metrics[0])-1):
      mlflow.log_metric(str(metrics[0][i +1]), values[i +1])
    mlflow.log_metric(str(metrics[0][i +1]), values[i +1])
    
  mlflow.end_run()
```

```
def save(log, runname, modelname, model, rouge_array, baseline): 
  #Save model metrics
  metrics = log_as_csv(log)
  
  if baseline == True:
    #Save RF model and metrics in MLFlow
    save_model_rf(runname, modelname, model, metrics, log, rouge_array)    
  else:
    #Save Neural Network model and metrics in MLFlow
    save_model(runname, modelname, model, metrics, log, rouge_array)
```
